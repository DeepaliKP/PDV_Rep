```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import csv
```

```python
user_agent = UserAgent()

#page_number=1

#main_url = 'https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DSamsung&page='+ str(page_number)

for i in range(5):
    #print(i)
    main_url = f'https://www.flipkart.com/search?sid=tyy%2C4io&otracker=CLP_Filters&p%5B%5D=facets.brand%255B%255D%3DSamsung&page={i+1}' 

    page = requests.get(main_url,headers={'user-agent':user_agent.chrome})
    soup = BeautifulSoup(page.content,'lxml')
    
    Model_name = []
    for name in soup.findAll("div", {"class": "_4rR01T"}):
        Model_name=Model_name+list(name)
        print(name.text)
    #a = name.text
    
        
    Model_price = []
    for name in soup.findAll("div", {"class": "_30jeq3 _1_WHN1"}):
        Model_price=Model_price+list(name)
        print(name.text)
    #b = name.text
    
        
    Model_ratings = []
    for name in soup.findAll("div", {"class": "_3LWZlK"}):
        Model_ratings=Model_ratings+list(name)
        print(name.text)
    #c = name.text
```

```python
final=list(zip(Model_name,Model_price,Model_ratings))
#final=list(zip(a, b, c))
filename='mob_Details.csv'
with open(filename, 'w', newline='', encoding="utf-8") as f:
    w=csv.writer(f)
    w.writerow(['Model_name','Model_price','Model_ratings'])
    for item in final:
        w.writerow(item)
    f.close()
```
